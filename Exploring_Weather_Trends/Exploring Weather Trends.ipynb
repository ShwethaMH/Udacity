```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# importing the data
global_temp = pd.read_csv('global_data.csv')
city_temp = pd.read_csv('city_data.csv')

# calculating the moving average
glb_mvg_avg = global_temp['avg_temp'].rolling(20).mean()
city_mvg_avg = city_temp['avg_temp'].rolling(20).mean()

# data visualizing using matplotlib
plt.plot(global_temp['year'], glb_mvg_avg, label = 'Global')
plt.plot(city_temp['year'], city_mvg_avg, label = 'Philadelphia')
plt.legend()
plt.xlabel("Years")
plt.ylabel("Temperature(Â°C)")
plt.title("Weather Trend")
plt.show()
```
